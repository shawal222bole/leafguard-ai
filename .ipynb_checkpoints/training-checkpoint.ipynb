```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
train_path = "training model/PlantVillage"
```

```python
print(os.listdir(train_path))
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)
```

```python
print(train_generator.class_indices)
```

```python
images, labels = next(train_generator)

plt.imshow(images[0])
plt.title(f"Label: {labels[0]}")
plt.axis("off")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(256,256,3)),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_generator,
    epochs=5
)
```

```python
plt.plot(history.history['accuracy'])
plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```

```python
os.makedirs("saved_models", exist_ok=True)

model.export("saved_models/1")

print("✅ MODEL SAVED")
```

```python
import os
import random
from PIL import Image
import numpy as np

# dataset path (your folder)
dataset_path = "training model/PlantVillage"

# get class folders
classes = os.listdir(dataset_path)

# pick random class
selected_class = random.choice(classes)

# go inside that class folder
class_path = os.path.join(dataset_path, selected_class)

# pick random image from that folder
image_name = random.choice(os.listdir(class_path))

# full image path
image_path = os.path.join(class_path, image_name)

print("Using image:", image_path)

# load image
img = Image.open(image_path).convert("RGB")
img = img.resize((256,256))

# preprocess
img_array = np.array(img)
img_array = img_array / 255.0
img_array = np.expand_dims(img_array, 0)

# predict
prediction = model.predict(img_array)

class_names = ['Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy']

print("Actual class:", selected_class)
print("Predicted:", class_names[np.argmax(prediction)])
print("Confidence:", np.max(prediction))
```
